## (예제 1) Json 파일읽기 (네이버증시)
json.loads() => json_normalize

json_normalize()함수는 중첩 된 JSON 문자열을 읽고 DataFrame을 반환하는 데 사용됨. 
이 함수를 사용하려면 먼저 Python의 JSON 라이브러리에 있는 json.loads()함수를 사용하여 JSON 문자열을 읽어야 함. 
그런 다음 이 JSON 객체를 json_normalize()에 전달하면 필요한 데이터가 포함된 Pandas DataFrame이 반환되는 형태

```python
import requests
import json
import pandas as pd
```

```python
headers = {'user-agent' :'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.67 Safari/537.36'}

dict_data = []
for i in range (0, 11) :
    url = f'https://finance.naver.com/world/worldDayListJson.naver?symbol=SPI@SPX&fdtc=0&page={i}'
    httpResponse = requests.get(url, headers=headers)
    dict_data.extend(json.loads(httpResponse.text))

form = pd.json_normalize(dict_data)
form
```
