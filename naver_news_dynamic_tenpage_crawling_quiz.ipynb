```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
```

```python
driver = webdriver.Chrome()
url = 'https://news.naver.com/main/main.naver?mode=LSD&mid=shm&sid1=101'

driver.get(url)
sleep(2)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')

title_list = []

def crawl_title() :
    global title_list
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    news_title = driver.find_elements(by=By.CSS_SELECTOR, value ='#section_body dt:nth-child(2) a')
    title_list += news_title

for pageloop in range(10) :
    crawl_title()
    driver.find_element(by=By.XPATH, value =f'//*[@id="paging"]/a[{pageloop+1}]')

for idx, title in enumerate(title_list ,1) :
    print(f'{idx}. {title.text}')
```
