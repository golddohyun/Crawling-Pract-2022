```python
from selenium import webdriver
from time import sleep
from bs4 import BeautifulSoup
import pandas as pd

driver = webdriver.Chrome()
url = 'https://www.youtube.com/feed/explore'
driver.get(url)

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
element = soup.select('#grid-container')

top50_title = element[1].select('#video-title > yt-formatted-string')
top50_name = element[1].select('#text > a')
top50_date = element[1].select('#metadata-line > span:nth-child(2)')
top50_date = element[1].select('#metadata-line > span:nth-child(1)')

youtube_table = []
for idx in range(0, len(top50_title)) :
    title = top50_title[idx].text
    name = top50_name[2*idx].text
    date = top50_date[idx].text
    views = top50_date[idx].text
    youtube_table.append([title, name, date, views])

youtube_result = pd.DataFrame(youtube_table, columns = ['제목', '게시자', '게시일', '조회수'])
youtube_result.to_csv("youtube_table.csv", mode='w')
print('<< 작업 완료 >>')
```
